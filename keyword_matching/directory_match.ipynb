```python
import pickle
with open('directory.pkl','rb') as f:
    directory = pickle.load(f)
```

```python
import re

def find_terms(text, directory):
    matches = []
    term_counts = {}
    
    for category, subcategories in directory.items():
        if isinstance(subcategories, dict):
            for subcategory, terms in subcategories.items():
                for term in terms:
                    for match in re.finditer(re.escape(term), text):
                        matches.append({
                            'topic': category,
                            'term': term,
                            'start_index': match.start(),
                            'end_index': match.end()
                        })
        else:
            for term in subcategories:
                for match in re.finditer(re.escape(term), text):
                    matches.append({
                        'topic': category,
                        'term': term,
                        'start_index': match.start(),
                        'end_index': match.end()
                    })
    for match in matches:
        term = match['term']
        if term in term_counts:
            term_counts[term] += 1
        else:
            term_counts[term] = 1

    # Add term counts to matches
    for match in matches:
        match['count'] = term_counts[match['term']]
    return matches
```

```python
result = find_terms(sample, directory)
for e in result:
    print(e)
```
