```python
# World Nuclear News: https://www.world-nuclear-news.org/
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time

driver = webdriver.Chrome()
driver.get('https://www.world-nuclear-news.org/search.aspx?searchtext=%22small+modular+reactor%22&searchmode=anyword')

time.sleep(1)

page = 1
with open('WNN_urls.txt', 'w') as file:
    
    TOT_PAGES = 53
    while (page < TOT_PAGES): # 52 pages in total

        for i in range(1,9):
            xpath = f'//*[@id="ctl00_ContentPlaceHolder1_PlaceHolderZones_lt_SearchResults_SmartSearchResults_srchResults_pnlSearchResults"]/div[{i}]/div/a'
            try:
                element = driver.find_element(By.XPATH, xpath)
            except NoSuchElementException:
                continue
            if (element):
                href = element.get_attribute('href')
                file.write(href + '\n')
                if i == 8:
                    print(element.text)
        
        try:
            if (page < 11):
                driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_PlaceHolderZones_lt_SearchResults_SmartSearchResults_srchResults_pnlSearchResults"]/a[13]').click()
            elif (page < 51):
                driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_PlaceHolderZones_lt_SearchResults_SmartSearchResults_srchResults_pnlSearchResults"]/a[14]').click()
            else:
                driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_PlaceHolderZones_lt_SearchResults_SmartSearchResults_srchResults_pnlSearchResults"]/a[5]').click()
        except NoSuchElementException:
            break
        
        #time.sleep(1)
        page += 1

time.sleep(5)
driver.quit()
```

```python
# New Civil Engineer: https://www.newcivilengineer.com/ 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time

driver = webdriver.Chrome()
driver.get('https://www.newcivilengineer.com/?s=%22small+modular+reactor%22')

#time.sleep(1)

page = 0
with open('NCE_urls.txt', 'w') as file:

    while(page <= 18): # 18 pages in total

        if page == 0:
            driver.find_element(By.XPATH, '//*[@id="top"]/div[11]/div[2]/div[1]/div[2]/div[2]/button[1]/p').click()

        
        for i in range(1,11):
            try:
                element = driver.find_element(By.XPATH, f'//*[@id="main-content"]/div[2]/div[3]/article[{i}]/h2/a')
            except NoSuchElementException:
                element = None

            #time.sleep(1)

            if (element):
                href = element.get_attribute('href')
                file.write(href + '\n')
                #print(href)
        
        try:
            driver.get(f'https://www.newcivilengineer.com/page/{page + 2}/?s=%22small+modular+reactor%22')
            print(f"Clicked {page +1} times")
        except NoSuchElementException:
            print("Nope")
            break
        
        #time.sleep(1)
        page += 1

driver.quit()
```

```python
# Nuclear Engineering International: https://www.neimagazine.com/
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time

driver = webdriver.Chrome()
driver.get('https://www.neimagazine.com/?s=%22small+modular+reactor%22')

time.sleep(3)

i = 1
with open('NEI_urls_5000.txt', 'w') as file:

    last_height = driver.execute_script("return document.body.scrollHeight")

    i=0
    while(i<5000): # Get full length

        try:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            if i%50 == 0:
                num_urls = 'iteration', i, 'total num urls', len(driver.find_elements(By.XPATH, "//a[@href and @title]"))
            i = i+1
        except NoSuchElementException:
            i = i+1
            break

    print("Scroll done. Sleeping for 5...")
    time.sleep(5)

    articles = driver.find_elements(By.XPATH, "//a[@href and @title]")
    for i, article in enumerate(articles):
        href = article.get_attribute('href')
        if 'news' in href or 'analysis' in href and i > 20:
            file.write(href + '\n')

driver.quit()
```
